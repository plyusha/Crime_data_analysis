```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import dateutil

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
nyc_rental_load = "Resources/Neighborhood_Zri_AllHomesPlusMultifamily.csv"
nyc_rental_data=pd.read_csv(nyc_rental_load)
nyc_mean_price_sales = "Resources/nyc_mean_price_per_sqft.csv"
nyc_price_per_sqft = pd.read_csv(nyc_mean_price_sales)
```

```python
nyc_rental_data.head()
```

```python
manhattan_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="New York"]
```

```python
manhattan_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
manhattan_rent_mean.dropna()
```

```python
bronx_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Bronx"]
bronx_rent_mean = bronx_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
bronx_rent_mean.dropna()
```

```python
queens_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Queens"]
queens_rent_mean = queens_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
queens_rent_mean.dropna()
```

```python
staten_island_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Richmond"]
staten_island_rent_mean = staten_island_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
staten_island_rent_mean.dropna()
```

```python
brooklyn_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Kings"]
brooklyn_rent_mean = brooklyn_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
brooklyn_rent_mean.dropna()
```

```python
df.dropna()
df_reset = df.reset_index()
```

```python
df_date = df_reset['Date'].str.split('-',expand=True)
```

```python
nyc_rental = df_reset.join(df_date, how='inner')
```

```python
nyc_rental1 = nyc_rental.dropna()
```

```python
nyc_rental1 = nyc_rental1.drop(columns=['index','Date'])
```

```python
df = nyc_rental1.rename(columns={0:'Year'})
```

```python
df['Year'] = df.Year.astype(int)
```

```python
df_by_year = df.groupby(['Year']).mean()
```

```python
df_by_year
```

```python
df_by_year_short = df_by_year.loc[df_by_year.index.isin(['2012','2013','2014','2015','2016','2017'])]
df_by_year_short.index
```

```python
df_by_year_short.plot(kind='line', title="RENTAL PRICES PER BOROUGH", legend=None).set(xlabel='Years', ylabel='Average Rental Prices')

plt.xticks(df_by_year_short.index.astype(int), list(df_by_year_short.index) )
#df_by_year_short.set(xlabel='Years')
```
